__Обработка столбцов с %__

```python
def percent_column_treatment(name_column, X):
    
    X[name_column] = X[name_column].map(lambda x: x[:-1])
    X[name_column] = X[name_column].replace('', float('nan'))
    X[name_column] = X[name_column].astype(float)
    
    return X
```

```python
for feature in percent_columns: 
    df = percent_column_treatment(feature, df)
```

```python
df.head()
```

__Обработка числовых столбцов__

```python
def number_column_treatment(name_column, X):
    
    X[name_column] = X[name_column].replace('-', float('nan'))
    X[name_column] = X[name_column].astype(float)
    
    return X  
```

```python
for feat in number_columns: 
    df = number_column_treatment(feat, df)
```

```python
df
```

__Обработка short_word_columns__

```python
def short_word_columns_treatment(name_column, X):
    list_column = X[name_column].to_list()
    f=[]
    for i in range(len(list_column)): 
        spam = list_column[i]
        if spam[-1] == 'B': 
            spam = float(list_column[i][:-1])*1*10**12
        elif spam[-1] == 'M': 
            spam = float(list_column[i][:-1])*1*10**6
        elif spam[-1] == 'K':
            spam = float(list_column[i][:-1])*1*10**3
        elif spam == '-':
            spam = float('nan')
        f.append(spam)
    X[name_column + '_full_count'] = f
    X = X.drop(name_column, axis=1)
    return X  
```

```python
for word_feat in short_word_columns: 
    df = short_word_columns_treatment(word_feat, df)
```
